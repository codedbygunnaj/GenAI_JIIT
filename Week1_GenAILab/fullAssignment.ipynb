```python
def userInput():
    textMessage=input("Enter the text you want to summarize: ")
    return textMessage
def promptGeneration(textMessage):
    promptMessage=f"Summarise the text: {textMessage}"
    return promptMessage
# if __name__=="__main__":
    # userMessage=userInput()
    # print(f"Original Message is: {userMessage} and prompt Generated is: {promptGeneration(userMessage)}")
    # promptGenerated=promptGeneration(userMessage)
```

```python
import google.generativeai as genai
genai.configure(api_key="")

def summarizeText(promptGenerated):
    model=genai.GenerativeModel("gemini-1.5-flash")
    response=model.generate_content(promptGenerated)
    return response.text

if __name__=="__main__":
        userMessage=userInput()
        promptGenerated=promptGeneration(userMessage)
        summary=summarizeText(promptGenerated)
        print(f"Summary is: {summary}")
```
