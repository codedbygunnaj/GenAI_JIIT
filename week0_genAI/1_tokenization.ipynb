```python
import tiktoken as tt
enc=tt.get_encoding("cl100k_base")
text="Generative AI is wonderFul"

token_ids=enc.encode(text)
print(f"Token ids of: {text} is: {token_ids}")
```

```python
for id in token_ids:
    word=enc.decode([id])
    print(f"This token_id ({id}) means {word}")
```
