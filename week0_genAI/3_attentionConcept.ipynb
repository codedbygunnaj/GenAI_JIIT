```python
word_meanings = {
    "bank": {
        "finance": "A place to keep money ðŸ’°",
        "nature": "Side of a river ðŸŒŠ"
    }
}
def attention_mechanism(sentence):
    words=sentence.lower().split()
    context="unknown" #default meaning

    if "money" in words or "deposit" in words:
        context="finance"
    elif "river" in words or "water" in words:
        context = "nature"

    if "bank" in words:
        print(f"Sentence: '{sentence}'")
        print(f"Context detected: '{context.upper()}'")
        print(f"Meaning of Bank: {word_meanings['bank'][context]}")

print("--- Test 1 ---")
attention_mechanism("I went to the bank to deposit money")
print("\n--- Test 2 ---")
attention_mechanism("I sat by the river bank")
```
