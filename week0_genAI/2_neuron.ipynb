```python
import torch
inputs=torch.tensor([1.0,1.0,1.0]) #rainy, play with friends, boss is at the office
#1.0 for YES and 0.0 for no

weights=torch.tensor([5.0,2.0,-10.0])
bias=torch.tensor(1) #assume mood good so +ve

outputs=torch.dot(inputs,weights)+bias

#activation function basics:
if outputs>0:
    print("HEHE, BUNK TIME")
else:
    print("CHUP CHAP KAAM KARO")
```
